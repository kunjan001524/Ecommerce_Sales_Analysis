# ***Step 3 – Data Merging (Master Dataset)***

### ***3.1: Merge Orders and Customers***

```python
master_df = pd.merge(
    orders,
    customers,
    on='customer_id',
    how='left'
)

print(master_df.shape)
```

**3.2 Merge Order items**

```python
master_df = pd.merge(
    master_df,
    order_items,
    on='order_id',
    how='left'
)

print(master_df.shape)
```

***3.3: Merge Products***

```python
master_df = pd.merge(
    master_df,
    products,
    on='product_id',
    how='left'
)

print(master_df.shape)
```

***3.4: Merge Sellers***

```python
master_df = pd.merge(
    master_df,
    sellers,
    on='seller_id',
    how='left'
)

print(master_df.shape)
```

***3.5: Merge Payments***

```python
# One order can have multiple payment records, so first aggregate the payments.

payments_summary = payments.groupby('order_id').agg({
    'payment_value': 'sum',
    'payment_installments': 'max',
    'payment_type': lambda x: ', '.join(x.unique())
}).reset_index()
```

```python
master_df = pd.merge(
    master_df,
    payments_summary,
    on='order_id',
    how='left'
)

print(master_df.shape)
```

***3.6: Merge Reviews***

```python
# One order can also have multiple review records, so summarize them first.

reviews_summary = reviews.groupby('order_id').agg({
    'review_score': 'mean'
}).reset_index()
```

```python
# Keep the first review for each order
reviews_summary = reviews[['order_id', 'review_score']].drop_duplicates(subset='order_id')

master_df = pd.merge(
    master_df,
    reviews_summary,
    on='order_id',
    how='left'
)

print(master_df.shape)
```

***3.7: Merge Category Translation***

```python
master_df = pd.merge(
    master_df,
    translation,
    on='product_category_name',
    how='left'
)

print(master_df.shape)
```

### ***Save the Master Dataset***

```python
master_df.to_csv("master_dataset.csv", index=False)

print("Master Dataset Created Successfully!")
print(master_df.shape)
```

# ***Step 4 – Sales Analysis***

## ***Question 1: Which Product Category Has the Highest Revenue?***

```python
category_revenue = master_df.groupby('product_category_name_english')['price'].sum().reset_index()

category_revenue = category_revenue.sort_values(
    by='price',
    ascending=False
)

category_revenue.head(10)
```

The highest revenue was generated by the top-ranked product category, indicating strong customer demand. This category contributes the largest share of sales and should be prioritized for inventory management and marketing campaigns.

## ***Question 2: Which Month Had the Highest Sales?***

```python
# Step 1: Create a Month Column

master_df['order_month'] = master_df['order_purchase_timestamp'].dt.to_period('M')
```

```python
# Calculate Monthly Sales

monthly_sales = master_df.groupby('order_month')['price'].sum().reset_index()

monthly_sales = monthly_sales.sort_values(by='price', ascending=False)

print(monthly_sales)
```

November 2017 recorded the highest sales revenue. This indicates strong customer demand during that month, which could be due to seasonal promotions or increased purchasing activity. The company should analyze the factors behind this performance and apply similar strategies in future campaigns.



## ***Question 3: Which Region Performs Best?***

```python
# Step 1: Calculate Revenue by State

region_sales = master_df.groupby('customer_state')['price'].sum().reset_index()
```

```python
# Step 2: Sort by Revenue

region_sales = region_sales.sort_values(by='price', ascending=False)

print(region_sales.head(10))
```

SP (São Paulo) generated the highest sales revenue, making it the best-performing region. The company should continue investing in this market while developing strategies to improve sales in lower-performing regions.

## ***Question 4: What is the Average Order Value (AOV)?***

```python
# Step 1: Calculate Total Order Value

order_value = master_df.groupby('order_id')['price'].sum().reset_index()
```

```python
# Step 2: Calculate the Average Order Value

average_order_value = order_value['price'].mean()

print("Average Order Value (AOV):", round(average_order_value, 2))
```

The Average Order Value (AOV) is ₹136.68 (or the equivalent currency in the dataset). This indicates the average amount customers spend per order. Increasing AOV through product bundles, cross-selling, and upselling can help improve overall revenue.

# **Step 5 - Data visulization**

### ***5.1 Bar Chart***

```python
# Bar Chart – Revenue by Product Category
# Set professional style
sns.set_theme(style="whitegrid")

# Top 10 categories by revenue
category_revenue = (
    master_df.groupby("product_category_name_english")["price"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

# Create figure
plt.figure(figsize=(14,7))

# Create bar chart
ax = sns.barplot(
    data=category_revenue,
    x="price",
    y="product_category_name_english",
    hue="product_category_name_english",
    palette="viridis",
    legend=False
)

# Add value labels
for i, value in enumerate(category_revenue["price"]):
    ax.text(
        value + 500,
        i,
        f"{value:,.0f}",
        va="center",
        fontsize=10,
        fontweight="bold"
    )

# Title and labels
plt.title(
    "Top 10 Product Categories by Revenue",
    fontsize=18,
    fontweight="bold",
    pad=20
)

plt.xlabel(
    "Total Revenue",
    fontsize=13,
    fontweight="bold"
)

plt.ylabel(
    "Product Category",
    fontsize=13,
    fontweight="bold"
)

plt.xticks(fontsize=11)
plt.yticks(fontsize=11)

# Remove top and right borders
sns.despine()

plt.tight_layout()
plt.show()
```

### ***5.2 Line Chart***

```python
# Monthly Sales Trend (High Quality)
# Professional theme
sns.set_theme(style="whitegrid")

# Create month column (if not already created)
master_df['order_month'] = master_df['order_purchase_timestamp'].dt.to_period('M').astype(str)

# Calculate monthly revenue
monthly_sales = (
    master_df.groupby('order_month')['price']
    .sum()
    .reset_index()
)

# Create figure
plt.figure(figsize=(14,7))

# Plot line chart
ax = sns.lineplot(
    data=monthly_sales,
    x='order_month',
    y='price',
    marker='o',
    linewidth=3,
    color='#1f77b4'
)

# Add value labels
for x, y in zip(monthly_sales['order_month'], monthly_sales['price']):
    plt.text(
        x,
        y,
        f"{y:,.0f}",
        ha='center',
        va='bottom',
        fontsize=9,
        fontweight='bold'
    )

# Title and labels
plt.title(
    "Monthly Sales Trend",
    fontsize=18,
    fontweight='bold',
    pad=20
)

plt.xlabel(
    "Month",
    fontsize=13,
    fontweight='bold'
)

plt.ylabel(
    "Total Revenue",
    fontsize=13,
    fontweight='bold'
)

# Rotate x-axis labels
plt.xticks(rotation=45)

# Remove top and right borders
sns.despine()

plt.tight_layout()
plt.show()
```

### ***5.4 Histogram Chart***

```python
# Histogram – Order Value Distribution
import matplotlib.pyplot as plt
import seaborn as sns

# Professional style
sns.set_theme(style="whitegrid")

# Calculate total order value for each order
order_values = (
    master_df.groupby('order_id')['price']
    .sum()
    .reset_index()
)

# Keep only the first 99% of data for better visualization
limit = order_values['price'].quantile(0.99)

filtered_orders = order_values[order_values['price'] <= limit]

# Average Order Value
avg_order = order_values['price'].mean()

# Create figure
plt.figure(figsize=(12, 7))

# Histogram
sns.histplot(
    data=filtered_orders,
    x='price',
    bins=30,
    kde=True,
    color='royalblue',
    edgecolor='black',
    alpha=0.8
)

# Average line
plt.axvline(
    avg_order,
    color='red',
    linestyle='--',
    linewidth=2,
    label=f'Average Order Value = {avg_order:.2f}'
)

# Titles and labels
plt.title(
    "Distribution of Order Values",
    fontsize=18,
    fontweight='bold'
)

plt.xlabel(
    "Order Value",
    fontsize=13,
    fontweight='bold'
)

plt.ylabel(
    "Number of Orders",
    fontsize=13,
    fontweight='bold'
)

plt.legend()
sns.despine()
plt.tight_layout()
plt.show()
```

### ***5.5 Pie Chart***

```python
# Pie chart for review score
# Professional style
sns.set_theme(style="whitegrid")

# Count review scores
review_counts = (
    master_df['review_score']
    .dropna()
    .astype(int)
    .value_counts()
    .sort_index()
)

# Attractive colors
colors = ['#FF6B6B', '#FFA94D', '#FFD43B', '#69DB7C', '#4D96FF']

# Highlight only 5-star reviews
explode = [0, 0, 0, 0, 0.08]

# Create figure
plt.figure(figsize=(8, 8))

plt.pie(
    review_counts,
    labels=[f"{i} Star" for i in review_counts.index],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    explode=explode,
    shadow=True,
    wedgeprops={
        'edgecolor': 'white',
        'linewidth': 2
    },
    textprops={
        'fontsize': 12,
        'fontweight': 'bold'
    }
)

plt.title(
    "Customer Review Score Distribution",
    fontsize=18,
    fontweight='bold',
    pad=20
)

plt.axis('equal')
plt.tight_layout()
plt.show()
```

### ***5.6 Heatmap***

```python
# Heatmap – Category vs Month (Revenue)
# Professional style
sns.set_theme(style="white")

# Create month column
master_df['order_month'] = pd.to_datetime(
    master_df['order_purchase_timestamp']
).dt.strftime('%b')

# Define month order
month_order = ['Jan','Feb','Mar','Apr','May','Jun',
               'Jul','Aug','Sep','Oct','Nov','Dec']

master_df['order_month'] = pd.Categorical(
    master_df['order_month'],
    categories=month_order,
    ordered=True
)

# Top 10 categories by revenue
top_categories = (
    master_df.groupby('product_category_name_english')['price']
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index
)

# Filter top categories
heatmap_df = master_df[
    master_df['product_category_name_english'].isin(top_categories)
]

# Create pivot table
pivot_table = heatmap_df.pivot_table(
    values='price',
    index='product_category_name_english',
    columns='order_month',
    aggfunc='sum',
    fill_value=0
)

# Plot heatmap
plt.figure(figsize=(14,8))

sns.heatmap(
    pivot_table,
    cmap='YlGnBu',
    annot=True,
    fmt='.0f',
    linewidths=0.5,
    linecolor='white',
    cbar_kws={'label': 'Revenue'}
)

plt.title(
    "Revenue by Product Category and Month",
    fontsize=18,
    fontweight='bold',
    pad=20
)

plt.xlabel(
    "Month",
    fontsize=13,
    fontweight='bold'
)

plt.ylabel(
    "Product Category",
    fontsize=13,
    fontweight='bold'
)

plt.tight_layout()
plt.show()
```
